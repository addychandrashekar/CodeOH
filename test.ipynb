```python

import matplotlib.pyplot as plt


samples_c1 = [(0, 0), (0, 1)]
samples_c2 = [(1, 0), (1, 1)]


C1_x, C1_y = zip(*samples_c1)
C2_x, C2_y = zip(*samples_c2)

# Create the plot
plt.figure(figsize=(6, 6))
plt.scatter(C1_x, C1_y, marker="o", color="blue", label="Class C1")
plt.scatter(C2_x, C2_y, marker="s", color="red", label="Class C2")

# Labels and legend
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Training Examples in Input Space")
plt.legend()
plt.grid(True)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)

# Show the plot
plt.show()
```

```python
import numpy as np


#  w = w(n) + η⋅(d[i]−y)⋅xi
def plr(x, w, d, eta, L):

    # Add bias term (column of ones) at the beginning of x
    x = np.hstack((np.ones((x.shape[0], 1)), x))

    for epochs in range(L):

        # Iterate over training samples
        for i in range(len(x)):
            # Compute weighted sum
            activation_potential = np.dot(w, x[i])

            # Activation function
            predicted_val = 1 if activation_potential >= 0 else -1

            # Update weights using correction factor (d[i] - y)
            w = w + eta * (d[i] - predicted_val) * x[i]

    return w

# x1  # x2  # x3  # x4
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  

# Desired outputs (d)
D = np.array([+1, +1, -1, -1])

# w(0) = (0, 0, 0)^T
w_initial = np.array([0.0, 0.0, 0.0]) 

# Learning rate
eta = 0.5
L = 3

# Run the Perceptron Learning Rule
print("Starting Perceptron Learning...")
final_w = plr(X, w_initial, D, eta, L)

print("\nFinal weight vector after training:", final_w)
```
